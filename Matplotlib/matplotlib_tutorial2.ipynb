```python
# Import matplotlib
from matplotlib import pyplot as plt
import numpy as np
import pandas as pd
```

```python
# Choosing plot style
plt.style.use('fivethirtyeight')

# Creating a basic pie chart using multiple iterables
slices = [60, 40]
slices2 = np.array(slices)
slices3 = {1:60, 2:40}
slices4 = tuple(slices)
slices5 = [[60, 40], [10, 70]]
slices6 = np.array(slices5)

plt.pie(slices)
plt.show()
plt.clf()

plt.pie(slices2)
plt.show()
plt.clf()


plt.pie(slices4)
plt.show()
plt.clf()

# plt.title('Example Pie Chart')
# plt.show()
```

```python
# Creating a series to store monopoly data
monopoly = pd.Series(np.array([400, 500, 1000, 300]), index=['Pash', 'Roman', 'Muvea', 'Shaddy'])
colors = ['#008fd5', '#fc4f30', '#e5ae37', '#6d904f']

# Creating pie chart of monopoly data
plt.pie(monopoly.array, labels=monopoly.index, wedgeprops={'edgecolor': 'black'}, colors=colors, shadow=True)

plt.title('Pie Chart of Monopoly Winnings')
plt.show()
plt.clf()
```

```python
# Stack plot  
# Track the contribution of player across the time of the game
minutes = np.arange(1, 10)
contribution = np.array([
    [1, 2, 3, 3, 4, 4, 4, 4, 5],
    [1, 1, 1, 1, 2, 2, 2, 3, 4],
    [1, 1, 1, 2, 2 ,2, 3, 3, 3]
])

plt.stackplot(minutes, contribution, labels=['Ronaldo', 'Messi', 'Neymar'])

plt.legend(loc='upper left')
plt.show()
plt.clf()
```

```python
# A list of ages and want to see their distribution
ages = [18, 19, 21, 25, 26, 26, 30, 32, 38, 45, 55]
bins = [10, 20, 30, 40, 50, 60]
plt.hist(ages, bins=bins, edgecolor='k')

plt.title('Distribution of Ages')
plt.ylabel('No. of respondents')
plt.xlabel('Ages')

plt.show()
plt.clf()
```

```python
# Scatterplots
x = [5, 7, 8, 5, 6, 7, 9, 2, 3, 4, 4, 4, 2, 6, 3, 6, 8, 6, 4, 1]
y = [7, 4, 3, 9, 1, 3, 2, 5, 2, 4, 8, 7, 1, 6, 4, 9, 7, 7, 5, 1]
colors = [7, 5, 9, 7, 5, 7, 2, 5, 3, 7, 1, 2, 8, 1, 9, 2, 5, 6, 7, 1]
sizes = [209, 486, 381, 255, 191, 315, 185, 228, 174, 538, 239, 394, 399, 153, 273, 293, 436, 501, 397, 539]

plt.style.use('seaborn')

plt.scatter(x, y, s=sizes, c=colors, cmap='Greens', edgecolor='black', alpha=0.75, linewidth=1)
cbar = plt.colorbar()
cbar.set_label('Satisfaction rating')

plt.tight_layout()
plt.show()
plt.clf()
```

```python
# Importing data
data = pd.read_csv('data.csv', header=0)

# Create a scatter plot showing relation between view_count and likes
plt.scatter(data.view_count, data.likes, c=data.ratio, cmap='summer', s=100, edgecolor='black', linewidth=1, alpha=0.75)

cbar = plt.colorbar()
cbar.set_label('Like/Dislike Ratio')

plt.yscale('log')
plt.xscale('log')

plt.title('Scatter plot of Views to likes of YT videos')
plt.xlabel('Likes')
plt.ylabel('View Count')

plt.show()
plt.clf()
```

```python
# Plotting time series data
from datetime import datetime, timedelta
from matplotlib import dates as mpl_dates


dates = [
    datetime(2019, 5, 23),
    datetime(2019, 5, 24),
    datetime(2019, 5, 26),
    datetime(2019, 5, 27),
    datetime(2019, 5, 28),
    datetime(2019, 5, 29),
    datetime(2019, 5, 30)
]

y = [0, 1, 3, 4, 6, 5, 7]

plt.plot_date(dates, y, linestyle='solid')
plt.gcf().autofmt_xdate()

date_format = mpl_dates.DateFormatter('%b, %d, %Y')
plt.gca().xaxis.set_major_formatter(date_format)


plt.show()
plt.clf()
```

```python
# Plot data from csv
data2 = pd.read_csv('data2.csv', parse_dates=['Date'])

plt.plot_date(data2.Date, data2.Close, linestyle='solid')

plt.gcf().autofmt_xdate()

plt.title('Close Price of Bitcoin against price')
plt.xlabel('Dates')
plt.ylabel('Prices in USD')
plt.tight_layout()

plt.show()
plt.clf()
```

```python
from itertools import count
from matplotlib.animation import FuncAnimation
import random

x_vals = []
y_vals = []

index = count()

def animate(i):
    global x_vals, y_vals
    x_vals.append(next(index))
    y_vals.append(random.randint(0, 6))
    
    plt.plot(x_vals, y_vals)

anim = FuncAnimation(plt.gcf(), animate, interval=1000)

plt.show()

```
